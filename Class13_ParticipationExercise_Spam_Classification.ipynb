```python
import nltk
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Take at least 10 sentences to train the model.
# Dataset
data = [
    ("Response now! You will miss a chance", "spam"),
    ("Urgent: Immediate action required to avoid suspension.", "spam"),
    ("You've won an iPhone! Click here immediately", "spam"),
    ("Urgent: Respond immediately before losing a chance to win.", "spam"),
    ("Congratulations! Claim your prize now", "spam"),
    ("Hello, I am a Data Analysis and Engineering course student", "not spam"),
    ("See you tomorrow at 11 AM in front of CCTB Campus.", "not spam"),
    ("Let's have lunch on Friday.", "not spam"),
    ("Reminder: Class 13 Participation Assignment due will be on Sunday.", "not spam"),
    ("Check out this Career Workshop.", "not spam"),
    ("Limited time offer: No-risk, high-return investment.", "spam"),
    ("Subscribe now for instant access to exclusive deals.", "spam"),
    ("Please review the document attached.", "not spam"),
    ("FedEx: Your package is on hold due to unpaid shipping fees.", "spam"),
]

# Split data into text and labels
texts = [text for text, label in data]
labels = [label for _, label in data]

# Preprocessing (Tokenization & Stopword Removal)
stop_words = set(stopwords.words('english'))

tokenized_texts = []
for text in texts:
    tokens = word_tokenize(text)  # Tokenize each sentence
    filtered = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]
    tokenized_texts.append(filtered)

# Convert List of Tokens back to String for CountVectorizer
processed_texts = [" ".join(tokens) for tokens in tokenized_texts]

# Feature Extraction using CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(processed_texts)
y = labels

# Split Data into Train and Test Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model Training
classifier = MultinomialNB()
classifier.fit(X_train, y_train)

# Prediction on Test Set
y_pred = classifier.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# 2.Predict the output/Class of a new text data
# New Text Prediction
new_processed_texts = vectorizer.transform( [
    "Limited time offer: Buy one, get one free!",
    "Do we have online classes today?",
    "Your prescription is ready for pickup at your preferred pharmacy.",
    "Secure your spot now! Limited tickets available.",
    "Please send me your revised resume for reviewing."
])
prediction = classifier.predict(new_processed_texts)
print(prediction)
```
