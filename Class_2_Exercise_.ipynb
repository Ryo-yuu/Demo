```python
x = df[['Private','Top10perc','S.F.Ratio','PhD','Terminal','perc.alumni','Room.Board','Outstate' ]]
```

```python
x.replace('Yes', 1, inplace=True)
x.replace('No', 0, inplace=True)
x
```

```python
x = x.drop(['PhD','Terminal'], axis=1)
```
