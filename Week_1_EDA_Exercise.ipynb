#Visualization

```python
df.value_counts('Wine')
```

```python
ax =sns.countplot(x='Wine',data=df);
ax.bar_label(ax.containers[0])
```

```python
# Get the first 5 unique values of 'Proline'
proline_values = df['Proline'].unique()[:5]

# Filter the DataFrame to include only these values
filtered_df = df[df['Proline'].isin(proline_values)]

# Create the countplot
sns.countplot(x='Proline', data=filtered_df)
plt.title('Count of Number of Each Proline (First 5)')
plt.show()
```

```python
# Print Pearson R correlation matrix
corr_matrix = df.corr(method='pearson')
corr_matrix
```

```python
# Sort the correlations from highest to lowest, for 'Proline' only.
corr_matrix[['Proline']].unstack().sort_values(ascending=False)
```

```python
# Visualize correlation matrix, which will show the correlations between the predictor variables with each other, and with the target ('Proline')
corr_matrix.style.background_gradient(cmap='coolwarm').format(precision=2)
```
