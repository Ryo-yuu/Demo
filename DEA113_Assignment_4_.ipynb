Data Reduction

```python
#Removing the Dizziness column
data = data.drop(['Dizziness'],axis=1)
```

```python
data = data.drop(['DifferenceLevel'],axis=1)
```

Creating Features

```python
data[['Age','Caffeine Intake (mg/day)','Stress Level (1-10)']]
```

```python
from datetime import datetime
current_year = datetime.now().year
data['Age'] = current_year - data['Age']
data.head()
```

```python
#Calculating the difference between Stress Level & Sweating Level and store into a new colum "DifferenceLevel(Stress/Sweating)"
data['DifferenceLevel(Stress/Sweating)'] = data['Stress Level (1-10)'] - data['Sweating Level (1-5)']
data.head()
```

```python
data.head(20)
```

```python
data['Caffeine Intake (mg/day)'] = data['Caffeine Intake (mg/day)'].fillna(data['Caffeine Intake (mg/day)'].mean())
print("Mean of Caffeine Intake (mg/day) is:", data['Caffeine Intake (mg/day)'].mean())
print("Median of Caffeine Intake (mg/day) is:", data['Caffeine Intake (mg/day)'].median())
print("Mode of Caffeine Intake (mg/day) is:", data['Caffeine Intake (mg/day)'].mode()[0])
```
